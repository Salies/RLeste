```python
from Pyleste.PICO8 import PICO8
from Pyleste.Carts.Celeste import Celeste
import Pyleste.CelesteUtils as utils
```

```python
def to_lua(sequence):
    return "[]" + ','.join([str(x) for x in sequence]) + ','
```

```python
p8 = PICO8(Celeste)
utils.load_room(p8, 0)
utils.skip_player_spawn(p8)
#utils.place_maddy(p8, 100, 0)
print(p8.game)
```

```python
fn = "./out/lvl0/best_episode_20250920_082542.txt"
sequence = open(fn).read().splitlines()
sequence = [int(x) for x in sequence]
for s in sequence:
    p8.set_btn_state(s)
    p8.step()
    print(p8.game)
```

```python
with open(fn.replace(".txt", ".tas"), "w") as f:
    f.write(to_lua(sequence))
```
